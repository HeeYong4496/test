```python
import pandas as pd
import numpy as np
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz' ,
                     'foo', 'foo', 'que', 'que'] ,
                    ['one', 'two', 'one',' two' ,     # two 앞에 스페이스가 하나 추가가 되었을때 아래에 미치는 영향
                     'one', 'two', 'one', 'two']]))

index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index =index, columns=['A', 'B'])
df2 = df[:4]
```

```python
df2
```

```python
stacked = df2.stack()
```

```python
stacked   # 스페이스 한칸 뛰고 two 출력
```

```python
stacked.unstack() # baz / two,one 순서가 바뀌어서 출력 (스페이스의 영향)
```

```python
stacked.unstack(0)  # 스페이스 영향으로 two 가 한번 더 적용
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)  # 기초 데이터에 스페이스 하나가 미치는 영향을 우연히 발견
```
